### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

np.random.seed(42)
sns.set(style="whitegrid")
print("Libraries loaded â€” Ready to learn!")
```

### Descriptive Statistics

```python
data = np.array([12,15,12,18,19,17,13,12,14,16])

print("Mean:", data.mean())
print("Median:", np.median(data))
print("Mode:", stats.mode(data).mode[0])
print("Variance:", data.var(ddof=1))
print("Std Dev:", data.std(ddof=1))
print("Min/Max:", data.min(), data.max())
print("IQR:", np.percentile(data,75) - np.percentile(data,25))
```
