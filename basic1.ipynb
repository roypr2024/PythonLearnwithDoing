### Counting VOWELS in a string

```python
vowels = ['a','e','i','o','u','A','E','I','O','U']
string = "Hi! I am learning Python."
connt = 0
for char in string:
    if char in vowels:
        connt += 1
print("Number of vowels:", connt)
```

### Count CONSONANTS in a given given word

```python
vowels = ['a','e','i','o','u','A','E','I','O','U']
string = "Hi! I am learning Python."
count = 0
for char in string:
    if char not in vowels and char.isalpha():
        count += 1
print("Number of consonants:", count)
```

### Count number of character occurring in a word

```python
string = "Hello, World!"
input=input("Enter a character to count its occurrences: ")
count = 0
for char in string:
    if char == input:
        count += 1
print(f"The character '{input}' occurs {count} times in the string.")
```

### Write a fibonacci series of n number

```python
first_num = 0
second_num = 1
range_number =  10

print("Fibonacci Series:")
for i in range(range_number):
    print(first_num)
    next_num = first_num + second_num
    first_num = second_num
    second_num = next_num
```

### Compairing two strings for ANAGRAMS
#### ANAGRAMS: An anagram is formed when two strings contain exactly the same characters with the same frequency, but possibly in a different order. For example, "listen" and "silent" are anagrams, while "hello" and "bello" are not.

### Using Sorting

#### Step 1: Check if both strings have the same length. If not, they cannot be anagrams. Step 2: Sort both strings alphabetically. Step 3: Compare the sorted results â€” if they match, the strings are anagrams.

```python
string1 = "hello"
string2 = "cello"
string1 = string1.replace(" ","").lower()
string2 = string2.replace(" ","").lower()
if len(string1) != len(string2):
    print("The strings are not anagrams (different lengths).")
else:
    sorted_str1 = sorted(string1)
    sorted_str2 = sorted(string2)
    print("Sorted String 1:", ''.join(sorted_str1))
    print("Sorted String 2:", ''.join(sorted_str2))
    if sorted_str1 == sorted_str2:
        print("The strings are anagrams.")
    else:
        print("The strings are not anagrams.")
```

### Counting Degits, Letters, and Spaces in a string

```python
import re # -- import reguler expression library
string1 = "Hello World 123 !"
digit_count = re.sub(r'[^0-9]', '', string1)
print("Number of digits in the string:", len(digit_count))
letter_count = re.sub(r'[^A-Za-z]', '', string1)
print("Number of letters in the string:", len(letter_count))
white_space_count = re.sub(r'[ \s]', '', string1)
print("Number of white spaces in the string:", len(string1) - len(white_space_count))
special_char_count = re.sub(r'[A-Za-z0-9\s]', '', string1)
print("Number of special characters in the string:", len(special_char_count))
```

### Building a Pyramid in pyhton

```python
def pyramid(n):
    for i in range(n):
        print(' ' * (n - i - 1) + '*' * (2 * i + 1))

pyramid(5)
```

```python
def pyramid(n):
    for i in range(n):
        for j in range(i,n):
            print(" ", end="")
        for j in range(i+1):
            print("*", end="")
        for j in range(i):
            print("*", end="")
        print("")

pyramid(5)
```
