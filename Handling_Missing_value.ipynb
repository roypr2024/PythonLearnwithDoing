## Missing Value
**Missing value** occurs in dataset when some of the informations is not stored for a variable.
There are 3 mechanisms

1. Missing Completly at Random (MCAR)
2. Missing at Random (MAR)
3. Missing Data not at Random (MNAR)

### Missing Completly at Random (MCAR)

**Missing completely at random (MCAR)** is a type of missing data machanism in which the probability of a value being missing is unrelated to both the observed data and the missing data. In other words, if the data is MCAR, the missing value are randomly distributed throughout the dataset, and **there is no systematic reason for why they are missing**.

**For Example**, in a survey about the prevalence of a certain disease, the missing data might be MCAR if the participants with missing values for certain questions were selected randomly and their missing responses are not related to their disease status or any other variables measured in the survey

### Missing at Random (MAR)

**Missing at random (MAR)** is a type of missing data machanism in which the probability of a value being missing depends only on the observed data, but not on the missing data itself. In other words, if the data is MAR, the missing value are systematically related to the observed data, but not to the missing data. 

**For Example**, Income Data: Suppose you are collecting income data form a group of people, but some participants choose not to report their income. If the decision to report or not report income is related to the participant's age or gender, but not to their income lavel, then the data is missing at random.

Medical Data: Suppose you are collecting medical data on patients, including their blood pressure, but some patients do not report their blood pressure. If the patients who do not report their blood pressure are more likely to be younger or have healthier lifestyles, but the missingness is not related to their actual blood pressure values, then the data is missing at random.

### Missing Data not at Random (MNAR)

**Missing Data not at Random (MNAR)** is a type of missing data mechanism where the probability of missing values depends on the value of the missing data itself. In other words, if the data is MNAR, the missingness is not random and is dependent on unobserved or unmeasured factors that are associated with the missing values.

**For Examples**, suppose you collecting data on the income and job satisfaction of employees in a company. If employees who are less satisfied with their jobs are more likelyto refuse to report their income, then the data is not missing at random. In this case, the missingness is dependent on job satisfaction, which is not directly observed or measured.

**df.dropna(axis=1)**, the parameter axis=1 specifies that the *operation should be performed along the columns*. The method will examine each column and, if a column contains any missing (NaN, None, or NaT) values, that *entire column will be removed from the DataFrame*. 

#### Explanation
1. axis=1 (or 'columns'): This tells pandas to operate horizontally across the data, which affects the columns.
2. dropna() method: This method is used to identify and remove missing values. By default, it removes items where any missing values are present (how='any').
3. Combined effect: df.dropna(axis=1) effectively means "drop all columns that contain at least one missing value." Only columns that are completely filled with non-missing data will remain in the resulting DataFrame. 

**df.dropna(axis=1, inplace=True)**, use where ***inplace=True***: This parameter tells pandas to modify the original DataFrame df directly, without creating a new DataFrame object. If inplace=False (the default), the method would return a new DataFrame with the columns dropped, and the original df would remain unchanged. 

```python
df.dropna(axis=1)
```
