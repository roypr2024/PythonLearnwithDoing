## Regression Mertics

**Depends on Context**
1.   MAE - Mean Absolute Error
2.   MSE - Mean Square Error
3.   RMSE - Rootover of MSE


**Independent on Context**
1.   R2 Score (*also known as coefficient of differentiation / goodness of fit*)
2.   Adjusted R2 Score

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
Y_pred = regressor.predict(X_test)
```

```python
Y_test.values
```

```python
mae = mean_absolute_error(Y_test, Y_pred)
print(f"Mean Absolute Error: {mae}")

mse = mean_squared_error(Y_test, Y_pred)
print(f"Mean Squared Error: {mse}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")

print(f"R2 Score: {r2_score(Y_test, Y_pred)}")

r2 = r2_score(Y_test, Y_pred)

k = 1  # number of independent variables
adjr2 = 1 - (((1-r2)*(X_test.shape[0] - 1))/(((X_test.shape[0] - 1)-1)-k))
print(f"Adjusted R2 Score: {adjr2}")
```

```python
class myLRClass:
  def __init__(self):
    self.m = None
    self.b = None

  def fit(self, X_train, Y_train):
    numerator = 0
    denominator = 0
    for i in range(X_train.shape[0]):
      numerator = numerator + ((X_train[i] - X_train.mean())*(Y_train[i] - Y_train.mean()))
      denominator = denominator + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))

    self.m = numerator/denominator
    self.b = (Y_train.mean() - (m * X_train.mean()))

    print(self.m)
    print(self.b)

  def predict(self, X_test):
    return self.m * X_test + self.b
```
